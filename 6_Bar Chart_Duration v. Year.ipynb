```python
import os
import pandas as pd
import matplotlib as plt
from datetime import datetime

file_path = ("Resources/ufo_clean.csv")

ufo_df = pd.read_csv(file_path)

ufo_df.head()
```

```python
#clean up df - only need datetime, state, duration (seconds)

#ufo_df["datetime"] = pd.to_datetime(ufo_df["datetime"], infer_datetime_format=True)
ufo_df["datetime"].dtype


```

```python
#rename duration column since it includes /n
ufo_df = ufo_df.rename(columns = {"duration/n(seconds)":"duration (seconds)"})
```

```python
#clean up dataframe; create new one with only necessary columns
new_ufo_df = ufo_df[["datetime","state","duration (seconds)"]]
new_ufo_df.head()
new_ufo_df["state"].count()
```

```python
#groupby state?
group_ufo_state = new_ufo_df.groupby(["state"]).sum()
group_ufo_state.head()
```

```python
duration_min = new_ufo_df["duration (seconds)"].min()
print(duration_min)

duration_max = new_ufo_df["duration (seconds)"].max()
print(duration_max)

duration_count = new_ufo_df["duration (seconds)"].value_counts()
print(duration_count)
```

```python
group_ufo_duration = new_ufo_df.groupby(["duration (seconds)"]).count()
group_ufo_duration.head()
```

```python
group_ufo_sighting = new_ufo_df.groupby(["datetime"]).count()
group_ufo_sighting.head()
```
