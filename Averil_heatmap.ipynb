```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import time
```

```python
file_path = ("Resources/ufo_clean.csv")
df = pd.read_csv(file_path)
df = df.rename(columns = {"latitude":"lat"})
df = df.rename(columns={ df.columns[-1]: "lng" })
df.head()
```

```python
state_df = df.groupby("state").count()
state_df = state_df.reset_index()
state_df = state_df.rename(columns = {"datetime":"count of state"})
state_df = state_df[["state", "count of state"]]
state_df.head()
```

```python
state_lat = df.groupby("state")["lat"].mean()
state_lat = state_lat.reset_index()
lat_df = pd.DataFrame(state_lat)
state_lng = df.groupby("state")["lng"].mean()
state_lng = state_lng.reset_index()
lng_df = pd.DataFrame(state_lng)

lat_df.head()
lng_df.head()
```

```python
coords = pd.merge(lat_df, lng_df, left_index=True, right_index=True)
coords = coords[["state_x", "lat", "lng"]]
coords.head()
```

```python
states = pd.merge(coords, state_df, left_index=True, right_index=True)
states = pd.DataFrame(states)
states.head()
```

```python
import gmaps
from config import (gkey)

gmaps.configure(api_key=gkey)
```

```python
# Store 'Lat' and 'Lng' into  locations 

locations = states[["lat", "lng"]].astype(float)

# Convert Poverty Rate to float and store

state_count = states["count of state"].astype(float)

# HINT: be sure to handle NaN values
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=state_count, dissipating=False, max_intensity=100, point_radius=1.25)
fig.add_layer(heat_layer)
fig
```
